### x or y  
x, y 둘 중 하나만 참이면 참

### x and y 
x, y 둘 다 참이어야 참

### not x
x가 거짓이면 참

```
cash = 10000
card = True
```

```
if cash >= 5000 or card:
  print('외식을 한다.')
else:
  print('집에서 먹는다.')
```

## 1-4.다양한 조건들은 elif
조건이 다양해지면 elif를 사용
<pre>
사용법:
if 조건문1:
  실행문1
elif 조건문2:
  실행문2
elif 조건문3:
  실행문3
else:
  실행문4

```
store = ['크림빵', '가위', '연필']
```

```
if '소보로' in store:
  print('소보로를 산다')
elif '크림빵' in store:
  print('크림빵을 산다')
else:
  print('상점을 나온다.')
```

### if문 한 줄로 작성하기

```
store = ['크림빵', '가위', '연필']
```

```
if '크림빵' in store:
  print('크림빵을 산다')
else:
  print('안산다')
```

```
# 한 줄로 작성
if '크림빵' in store: print('크림빵을 산다')
```
