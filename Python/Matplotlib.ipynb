```
# 그럼 우리가 원하는 위치에 그림을 그리기 위해서는?
# 위에서 add_subplot 을 할 때 변수명을 지정하여 반환값을 받았다.
# 해당 변수를 통해 plot을 그리면 된다.

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.randn(100), bins=20)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
```

```
# battles 데이터셋에는 각 공격별로 공격군과 수비군에 대한 정보가 있음
# 공격군과 수비군의 병사수도 있음.
# attacker_size, defender_size
# 공격군과 수비군의 수가 1만명이 넘어가는 대규모 전투 상의 공/수군의 병력차이를 확인해보고
# box plot으로 시각화 해보자.
```

```
# 전투이름을 인덱스로 지정하기
battles = battles.set_index(['name'])
battles.head()
```

```
# 전체 병력(공/수 합)이 1만명을 넘어가는 행 선택
# boolean mask 만들기
large_battle_mask = battles['attacker_size'] + battles['defender_size'] > 10000
```

```
# 이들의 병력 정보만 추출하여 변수에 담기
# large_battles 데이터프레임 만들기, 열은 attacker_size, defender_size를 가지게.
large_battles = battles.loc[large_battle_mask, ['attacker_size', 'defender_size']]
```

```
large_battles.shape
```

```
large_battles
```

```
# 수평 barplot을 그리고 stacked옵션을 넣어줌.
# 10개의 대규모 전투에 대해 공/수 병력 차이를 나타내는 바플롯을 그림
ax2 = large_battles.plot(kind='barh', stacked=True, fontsize=8)
```

```
# 단일전투 하나하나의 공수 상대적 비율이 잘 들어나지 않았음.
# 각 전투에서의 병력의 상대적 비율을 알고자 함
# large_battles 데이터프레임에 새로운 컬럼 추가
# 공격군/공+수군 : 공격군의 비율
large_battles['attacker_pcts'] = large_battles['attacker_size']/(large_battles['attacker_size'] + large_battles['defender_size'])
```

```
# 수비군의 비율
large_battles['defender_pcts'] = large_battles['defender_size']/(large_battles['attacker_size']+large_battles['defender_size'])
```

```
large_battles
```

```
# 두개 열로 바플롯 그리기
# 각 전투 별 공수군 간의 병력차이가 확연히 드러나게 보일 수 있다.
ax3 = large_battles[['attacker_pcts', 'defender_pcts']].plot(kind='barh', stacked=True, fontsize=8)
```
